```python
import re
pattern_1 = r'จบการศึกษาจาก : +'
pattern_2 = r'(w)*งานปัจจุบัน : +'
education_list = []
career_list = []
num_1,num_2 = 0,0

for num,i in enumerate(w.find_all('p')):
  if re.findall(pattern_1,i.text) != []:
    text = i.text.replace('จบการศึกษาจาก : ','')
    text = re.sub('(.)*– ','',text)
    num_1 = num
    education_list.append(text)
  if re.findall(pattern_2,i.text) != []:
    text = i.text.replace('งานปัจจุบัน : ','')
    text = text.replace('สาย','')
    num_2 = num
    career_list.append(text)
  if num_2-num_1 > 1:
    education_list.pop(-1)
    edu_text = [re.sub('(.)*– ','',a.text) for a in w.find_all('p')[num_1:num_2]]
    edu_text = ', '.join(edu_text)
    education_list.append(edu_text)
print(education_list)
print(career_list)
    
  
```

```python
img = w.find_all('img')
pattern = r'(.)*เรียนไม่ตรงสาย(.)*'
img_list = [img_tag['src'] for img_tag in img if re.findall(pattern,img_tag['src']) !=[] ]
img_list
```

```python
img = w.find_all('img')
pattern_img = r'(.)*เรียนไม่ตรงสาย(.)*'
img_list = [img_tag['src'] for img_tag in img if re.findall(pattern_img,img_tag['src']) !=[] ]
img_list
```

```python
pattern_comment = r'“(.)*'
comment=w.find_all('p')
comment_list = [comment_text.text for comment_text in comment if re.findall(pattern_comment,comment_text.text) !=[] ]
comment_list
```

```python
import json
data = [{'name' : name[i],
        'education' : education_list[i],
        'career' : career_list[i],
        'comment' : comment_list[i],
        'image' : img_list[i]
        } for i in range(len(name))]

jason_data = json.dumps(data)

with open('มอง_7_มุม_จาก_7_คนจบไม่ตรงสาย.json', 'w') as outfile:
    outfile.write(jason_data)
```

```python
df = pd.DataFrame.from_dict(data)
df.to_parquet("มอง_7_มุม_จาก_7_คนจบไม่ตรงสาย.json.parquet")
```
